#### After checking the Columns, we found out that the dataset is in wide format, having years(eg. 2020,2021, etc) as columns, now we have to covert the datasets into Long format

```python
# Example for Dataset 1 (Cost of a healthy diet)
data1_long = data1_about_sa.melt(
    id_vars=['REF_AREA_LABEL'],  # keep country name
    value_vars=[str(y) for y in range(2017, 2023)],  # years available in dataset 1
    var_name="Year",
    value_name="Cost_Healthy_Diet"
)

# Dataset 2 (GDP per capita)
data2_long = data2_about_sa.melt(
    id_vars=['REF_AREA_LABEL'],
    value_vars=[str(y) for y in range(2000, 2025)],
    var_name="Year",
    value_name="GDP_per_capita"
)

# Dataset 3 (Undernourished)
data3_long = data3_about_sa.melt(
    id_vars=['REF_AREA_LABEL'],
    value_vars=[str(y) for y in range(2002, 2025)],
    var_name="Year",
    value_name="Undernourished"
)

# Dataset 4 (Unable to afford healthy diet)
data4_long = data4_about_sa.melt(
    id_vars=['REF_AREA_LABEL'],
    value_vars=[str(y) for y in range(2017, 2025)],
    var_name="Year",
    value_name="Unable_to_Afford"
)
```

#### Giving the year column a correct data type

```python
# Convert year column to integer
for df in [data1_long, data2_long, data3_long, data4_long]:
    df["Year"] = df["Year"].astype(int)
```

#### Now that we have filtered our datasets to focus only on South Africa, we now have to merge them together to make one valuable dataset.

```python
# Merge datasets
merged = data1_long.merge(data2_long, on=["REF_AREA_LABEL","Year"], how="outer")
merged = merged.merge(data3_long, on=["REF_AREA_LABEL","Year"], how="outer")
merged = merged.merge(data4_long, on=["REF_AREA_LABEL","Year"], how="outer")

# Preview
merged.head(15)
```

```python
merged.tail(15)
```

#### After merging, we found out that some of the columns, (Cost Of Healthy Diet & Unable to Afford) have data starting from 2017 and years before are have not been recorded, as of we cannot predict the data going backwards, we then dropped the rows, which values are null until 2016 and then filling the Cost Of Healthy Diet 2023 & 2024 with interpolation and estimating the values based on the previsous trend 

```python
# Filling Cost_Healthy_Diet 2023 & 2024 with interpolation and estimating the values based on the previsous trend
merged["Cost_Healthy_Diet"] = merged["Cost_Healthy_Diet"].interpolate(method="linear")
```

```python
merged.tail()
```

```python
# Dropping Null Values. New Dataset
merged_clean_data = merged.dropna()
```

```python
merged_clean_data
```

#### Now that we have filled the missing values, our dataset has to be in good format to be able to tell a story.

#### Now we have to rename our columns by giving them descriptive names, to help with clarity and readability, show consistency across datasets, and error reduction.

```python
#Renaming of columns for consistency
merged_clean_data.rename(columns={'REF_AREA_LABEL': 'Country Name'}, inplace = True)
merged_clean_data                      
```

```python
#Renaming of columns for consistency
merged_clean_data.rename(columns={'Cost_Healthy_Diet': 'Cost Of Healthy Diet(PPP$)'}, inplace = True)
merged_clean_data.rename(columns={'GDP_per_capita': 'GDP Per Capita (constant 2017 international $)'}, inplace = True)
merged_clean_data.rename(columns={'Undernourished': 'Undernourished People (million)'}, inplace = True)
merged_clean_data.rename(columns={'Unable_to_Afford': 'People Unable To Afford Healthy Diet (million)'}, inplace = True)
```

#### Here we are generating new columns GDP Per Day and Affordability Ratio. These are our derived indicators which will help us show relationships between the diet cost, income, and affordability of food in South Africa of the course of 8 years(2017 -2024).

```python
# Since GDP Per Capita is given per year, we have to calculate the GDP per day in order to find the cost of the healthy diet per day.
merged_clean_data["GDP_Per_Day"] = merged_clean_data["GDP Per Capita (constant 2017 international $)"] / 365
```

```python
merged_clean_data["GDP_Per_Day"]
```

```python
# Calculating the affordability ratio measures thhe share of daily income needed for a healthy diet.
merged_clean_data["Affordability_Ratio"] = merged_clean_data["Cost Of Healthy Diet(PPP$)"] / merged_clean_data["GDP_Per_Day"]
```

```python
merged_clean_data["Affordability_Ratio"]
```

```python
merged_clean_data[ ["GDP_Per_Day", "Affordability_Ratio"]].describe()
```

##### The table presents descriptive statistics for two derived indicators: GDP per Day and the Affordability Ratio. 
##### On average, South Africans had a daily income of 38.2 PPP dollars per person, with a range between 36.3 and 40.0 PPP $/day across 2017–2024. 
##### When compared against the cost of a healthy diet, the Affordability Ratio averaged 0.088 (≈8.9%), meaning that a healthy diet required less than 10% of daily income on average.

##### While this suggests affordability at a national level, the relatively small variation in the ratio (std = 0.01) highlights that the challenge is not national food prices but income inequality. 
##### The ratio appears reasonable in aggregate, yet nearly 37 million South Africans remain unable to afford a healthy diet, illustrating the gap between averages and lived reality. 

## Step 2, Numpy Calculations

```python
#Extracting arrays from the dataframe
years = merged_clean_data["Year"].values
cost = merged_clean_data["Cost Of Healthy Diet(PPP$)"].values
gdp = merged_clean_data["GDP Per Capita (constant 2017 international $)"].values
unable = merged_clean_data["People Unable To Afford Healthy Diet (million)"].values

# % year-on-year change in cost
#This calculates how much the cost of healthy diet changes each year as a percentage.
pct_change_cost = np.diff(cost) / cost[:-1] * 100
print("Year-on-year % change in cost:", pct_change_cost)


# Moving average (3-year)
#Smooths out short-term fluctuations to identify longer-term trends in cost data.
ma_cost = np.convolve(cost, np.ones(3)/3, mode="valid")
print("3-year moving avg cost:", ma_cost)
```

## Step 3 Visualization

```python
sns.set(style="whitegrid", palette="deep")

def plot_series(y, title, ylabel):
    plt.figure(figsize=(9,5))
    sns.lineplot(data=merged_clean_data, x="Year", y=y, marker="o", color="blue")
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel("Year")
    plt.tight_layout()
    plt.show()

plot_series("GDP Per Capita (constant 2017 international $)","GDP per Capita (PPP, constant 2017 $) - South Africa","2017 Int. $")
plot_series("Cost Of Healthy Diet(PPP$)","Cost of a Healthy Diet (PPP $/day) - South Africa","PPP $/day")
plot_series("Affordability_Ratio","Affordability Ratio - South Africa","Share of daily income")
plot_series("Undernourished People (million)","Undernourished Population - South Africa","Millions")
plot_series("People Unable To Afford Healthy Diet (million)","Unable to Afford Healthy Diet - South Africa","Millions")

# Compare undernourished vs unable to afford
plt.figure(figsize=(9,5))
sns.lineplot(data=merged_clean_data, x="Year", y="Undernourished People (million)", marker="o", label="Undernourished")
sns.lineplot(data=merged_clean_data, x="Year", y="People Unable To Afford Healthy Diet (million)", marker="o", label="Unable to Afford")
plt.title("Undernourished vs Unable to Afford Healthy Diet - South Africa")
plt.ylabel("Millions")
plt.xlabel("Year")
plt.legend()
plt.tight_layout()
plt.show()
```

### GDP per Capita (PPP, constant 2017 $) - South Africa

##### Based on our analysis, South Africa’s GDP per capita remained relatively stable between 2017 and 2019, but experienced a sharp decline in 2020, largely due to the economic disruptions caused by the COVID-19 pandemic. 
##### Although there was a partial recovery in 2021–2022, the momentum was not sustained, and by 2023–2024 GDP levels had once again declined. 
##### This trend points to a fragile and uneven economic recovery, where external shocks and structural economic weaknesses continue to undermine long-term stability.
##### The decline in GDP per capita has a direct impact on household purchasing power, making it increasingly difficult for families to keep pace with rising food costs.

### Cost of a Healthy Diet (PPP $/day) - South Africa

##### The graph indicates that the cost of a healthy diet was relatively low from 2017 - 2024 ranging from 3.05 - 3.74 PPP dollers per day, with an average of 3.88 PPP dollers per day. This reflects a fairly stable cost with a slight growth(about 20 cents increase from median to max), suggesting that healthy diet prices have not skyrocketed in this period.
##### Food price stability might reflect controlled inflation or relatively steady agricultural output, but even small increasesncan stress low-income households.

### Affordability Ratio - South Africa
##### The graph measures the share of a daily income needed by the average South African be able to afford a healthy diet. The ratio ranges for 0.0076 - 0.100 with an average 0.088(~ 9%). This suggests that on average, a healthy diet cost about 9% of daily income (FAO often uses 52% of daily income as a cutoff for extreme food unaffordability).
##### The ratio is below 10%, but 37 million people still cannot afford a healthy diet which shows the income inequality gap: the national average affordability does not reflect the reality of poorer households.



### Undernourished Population - South Africa

##### The line chart shows that undernourishment in South Africa worsened, rising from 3.5M in 2017 to 6.3M in 2024 with an average 4.8 million people lacking access to nutritious food. Even though GDP per capita rose slightly, nutrition outcomes declined - this suggests income inequality, food distribution issues or the economic shocks( e.g the effect of COVID-19 inflation) preveting growth from reaching vulnerable groups.
##### This highlights that economic growth during this period did not translate into improved food security for vulnerable groups.

### Unable to Afford Healthy Diet - South Africa

##### The graph indicates that 35–39.5 million South Africans were unable to afford a healthy diet between 2017–2024, increasing by 4.5M over time with  an average of 37.2 million. This far outnumbers the 3–6 million South Africans who were undernourished; roughly two-thirds of South Africa’s population (≈60 million total) cannot afford a healthy diet. 
##### This demonstrates that while many households consume enough calories to avoid hunger, the quality of diets is compromised, reflecting affordability rather than absolute food scarcity.

### "Undernourished vs Unable to Afford Healthy Diet - South Africa"

##### The number of people unable to afford a healthy diet are approx. 37M and population of undernourished people is approx. 5M.  Comparing the two trends reveals a consistent gap: the population unable to afford healthy diets is six to eight times higher than those classified as undernourished. 
##### This means that most South Africans can get calories but not nutritious diets. The gap points to diet quality issues rather than just hunger.

```python
sns.set(style="whitegrid", palette="deep")

# --- Bar Chart: GDP in selected years ---
plt.figure(figsize=(9,5))
df_subset = merged_clean_data[merged_clean_data["Year"].isin([2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024])]  # example years
sns.barplot(data=df_subset, x="Year", y="GDP Per Capita (constant 2017 international $)", color="skyblue")
plt.title("GDP per Capita over 8 years - South Africa")
plt.ylabel("2017 Int. $")
plt.xlabel("Year")
plt.tight_layout()
plt.show()

# --- Area Chart: Undernourished population ---
plt.figure(figsize=(9,5))
plt.fill_between(merged_clean_data["Year"], merged_clean_data["Undernourished People (million)"], alpha=0.4, color="red")
plt.plot(merged_clean_data["Year"], merged_clean_data["Undernourished People (million)"], marker="o", color="red")
plt.title("Undernourished Population (Millions) - South Africa")
plt.ylabel("Millions")
plt.xlabel("Year")
plt.tight_layout()
plt.show()

# --- Scatter Plot: GDP vs Affordability Ratio ---
plt.figure(figsize=(9,5))
sns.scatterplot(data=merged_clean_data, x="GDP Per Capita (constant 2017 international $)", y="Affordability_Ratio",
                size="People Unable To Afford Healthy Diet (million)", hue="Year", palette="viridis", alpha=0.8)
plt.title("GDP per Capita vs Affordability Ratio (Bubble = Population Unable to Afford)")
plt.xlabel("GDP per Capita (PPP, 2017 Int. $)")
plt.ylabel("Affordability Ratio")
plt.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# --- Histogram: Distribution of Affordability Ratio ---
plt.figure(figsize=(9,5))
sns.histplot(merged_clean_data["Affordability_Ratio"], bins=10, kde=True, color="green")
plt.title("Distribution of Affordability Ratio - South Africa")
plt.xlabel("Share of Daily Income")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

### Undernourished Population (Millions) - South Africa

##### Even small changes in GDP per capita or food costs can significantly affect millions of people.


### GDP per Capita vs Affordability Ratio (Bubble = Population Unable to Afford)

##### The scatter plot shows that higher GDP per capita generally corresponds to a lower affordability ratio. Larger bubbles (meaning more people unable to afford healthy diets) appear when affordability ratios are high, despite modest GDP growth.
